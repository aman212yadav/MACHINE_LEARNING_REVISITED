```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y1=x**2
y2=x**2+4
```

```python
plt.style.use('dark_background')
plt.plot(x,y1,label='line 1')
plt.plot(x,y2,label='line 2')
plt.title('data visualization ')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```
