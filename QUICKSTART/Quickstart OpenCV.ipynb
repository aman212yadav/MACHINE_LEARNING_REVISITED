```python
import cv2
import matplotlib.pyplot as plt
```

```python
grey_img=cv2.imread('aman212yadav.jpg')
print(grey_img.shape)
```

```python
plt.imshow(grey_img)
plt.axis('off')
plt.title('BGR_AMAN '+str(grey_img.shape))
plt.show()
```

```python
rgb_img=cv2.cvtColor(grey_img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
plt.axis('off')
plt.title('RGB_AMAN '+str(rgb_img.shape))
plt.show()
```

```python
resized_img=cv2.resize(rgb_img,(100,100))
```

```python
plt.imshow(resized_img)
plt.axis('off')
plt.title('RGB_AMAN '+str(resized_img.shape))
plt.show()
```
