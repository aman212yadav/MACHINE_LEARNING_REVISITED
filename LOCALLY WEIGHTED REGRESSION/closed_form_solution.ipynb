```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
# generating dataset
X,Y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=3,random_state=10)
```

```python
print(X.shape)
Y=Y.reshape((-1,1))
print(Y.shape)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
# Normalisize
X=(X-X.mean())/X.std()
plt.scatter(X,Y)
plt.title('Normalised data')
plt.show()
```

```python
ones=np.ones((X.shape[0],1))
```

```python
X=np.hstack((X,ones))
```

```python
print(X.shape)
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def getThetaClosestForm(X,Y):    # formula theta=( (X.T X).T)X.T Y
    a=np.dot(X.T,X)
    b=np.dot(X.T,Y)
    theta=np.linalg.pinv(a)
    theta=np.dot(theta,b)
    return theta
theta=getThetaClosestForm(X,Y)
print(theta)
print(theta.shape)
```

```python
print(X.shape,Y.shape)
plt.scatter(X[:,0],Y)
plt.title('Normalised data')
plt.plot(X[:,0],predict(X,theta),color='red')
plt.show()
```
