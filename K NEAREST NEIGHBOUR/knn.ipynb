```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading data
dfx=pd.read_csv('DATASET/xdata.csv',header=None,index_col=0)
X=dfx.values
dfy=pd.read_csv('DATASET/ydata.csv',header=None,index_col=0)
Y=dfy.values
Y=Y.reshape((-1,))
```

```python
# visualising data
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
# knn algorithm


# distance
def distance(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))


def knn(X,Y,query_point,k=6):
    m=X.shape[0]
    l=[]
    for i in range(m):
        dist=distance(query_point,X[i])
        l.append((dist,Y[i]))
    val=sorted(l)
    val=val[:k]
    val=np.array(val)
    r=np.unique(val[:,1],return_counts=True)
    idx=np.argmax(r[1])
    return r[0][idx]
    
   
```

```python
knn(X,Y,np.array([4,5]))
```
