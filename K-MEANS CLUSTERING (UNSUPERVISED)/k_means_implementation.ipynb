```python
# importing libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
```

```python
# generating clusters
X,Y=make_blobs(n_samples=500,n_features=2,centers=5)
```

```python
print(X.shape,Y.shape)
```

```python
# plotting clusters
plt.scatter(X[:,0],X[:,1],c=Y)
plt.grid()
plt.title("data")
plt.show()
```

```python
k=5
colors=['green','red','blue','yellow','orange']
clusters={}
for i in range(k):
    centre=10*(2*np.random.random((X.shape[1],))-1)
    points=[]
    cluster={
        'centre':centre,
        'points':points,
        'color':colors[i]  
    }
    clusters[i]=cluster
    
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def assignPointsToCluster(clusters):
    print(X.shape[0])
    for ix in range(X.shape[0]):
        dist=[]
        for j in range(k):
            dist.append(distance(X[ix],clusters[j]['centre']))
        group=np.argmin(dist)
        clusters[group]['points'].append(X[ix])
```

```python
def updateClusters(clusters):
    for i in range(k):
        pts=np.array(clusters[i]['points'])
        if pts.shape[0]>0:
            u=np.mean(pts,axis=0)
            clusters[i]['centre']=u
            clusters[i]['points']=[]
        

def plotClusters(clusters):
    for i in range(k):
        pts=np.array(clusters[i]['points'])
        print(pts.shape)
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[i]['color'])
        except:
            pass
        centre=clusters[i]['centre']
        print(centre[0])
        plt.scatter(centre[0],centre[1],marker="*",c='black')
```

```python
assignPointsToCluster(clusters)
plotClusters(clusters)
updateClusters(clusters)
```
