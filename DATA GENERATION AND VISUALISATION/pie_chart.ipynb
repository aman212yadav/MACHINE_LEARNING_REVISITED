```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels=['math','physics','chemistry','english']
size=[30,30,25,15]
plt.pie(size,labels=labels,autopct='%1.2f%%',explode=[0.1,0,0.1,0],shadow=True)
plt.plot()
```
