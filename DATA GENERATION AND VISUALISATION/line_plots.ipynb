```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
y1=x**2
y2=2*x+3
```

```python
plt.style.use(['dark_background','seaborn'])
```

```python
plt.plot(x,y1,color='red',label='label1',marker='o')
plt.plot(x,y2,color='green',label='label2',linestyle='dashed')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('this is title',color='orange')
plt.legend()
plt.show()
```
