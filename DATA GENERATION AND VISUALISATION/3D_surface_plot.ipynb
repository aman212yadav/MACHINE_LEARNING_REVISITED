```python
import numpy as np
import mpl_toolkits.mplot3d as Axes3D
import matplotlib.pyplot as plt
```

```python
a=np.arange(-4,4,0.02)
b=np.arange(-4,4,0.02)
a,b=np.meshgrid(a,b)
```

```python
#print(a)
#print(b)
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2)
plt.show()
```

```python
plt.contour(a,b,a**2+b**2)
plt.show()
```
