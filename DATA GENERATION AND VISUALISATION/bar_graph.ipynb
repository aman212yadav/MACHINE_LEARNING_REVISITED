```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.ylim(0,30)
plt.style.use(['seaborn','dark_background'])
x=np.array([1,3,5])
plt.bar(x,[10,20,5],width=0.5,label='current year',tick_label=['gold','silver','platinum'],color='orange')
plt.bar(x+0.5,[5,15,10],width=0.5,label='next year',color='green')
plt.xlabel('Metal',color='red')
plt.ylabel('Price',color='pink')
plt.title('Metal price ')
plt.legend()
plt.show()
```
