# Gradient Descent
- start with a random theta
  - Update theta according to the rule

```python
#  hypothesis
def hypothesis(X,theta):
    return theta[0]+theta[1]*X

# error 
def error(X,Y,theta):
    m=X.shape[0]
    error=0
    for i in range(m):
        hx=hypothesis(X[i],theta)
        error+=(hx-Y[i])**2
    return error

# gradient calculation
def grad(X,Y,theta):
    gra=np.zeros((2,))
    m=X.shape[0]
    for i in range(m):
        hx=hypothesis(X[i],theta)
        gra[0]+=(hx-Y[i])
        gra[1]+=(hx-Y[i])*X[i]
    return gra    

# gradient descent
def gradient_descent(X,Y,learning_rate=0.001):
    theta=np.zeros((2,))
    max_iteration=100
    erro=[]
    i=0
    while i < max_iteration:
        err=error(X,Y,theta)
        erro.append(err)
        gra=grad(X,Y,theta)
        theta[0]=theta[0]-learning_rate*gra[0]
        theta[1]=theta[1]-learning_rate*gra[1]
        
        i+=1
    return erro,theta

error,theta=gradient_descent(X,Y)
print(theta)
```

```python
X_test=np.linspace(-4,6,10)
```

```python
plt.scatter(X,Y)
plt.plot(X_test,hypothesis(X_test,theta),color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
