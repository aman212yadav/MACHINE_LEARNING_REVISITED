```python
import cv2
import matplotlib.pyplot as plt
```

```python
# read 
mustache             =  cv2.imread('DATASET/Train/mustache.png', -1)
glasses              =  cv2.imread('DATASET/Train/glasses.png' , -1)

im=cv2.imread('DATASET/Train/Jamie_Before.jpg',-1)
grey_img=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
grey_im=cv2.cvtColor(im,cv2.COLOR_BGR2BGRA)
im=cv2.cvtColor(im,cv2.COLOR_BGR2BGRA)
```

```python
plt.imshow(im)
plt.show()
plt.imshow(grey_img)
plt.show()
```

```python
# plotting using cv2 libraries
cv2.imshow(' DOG ',im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
nose_classifier=cv2.CascadeClassifier('third_party/Nose18x15.xml')
eye_classifier=cv2.CascadeClassifier('third_party/frontalEyes35x16.xml')
face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
face=face_classifier.detectMultiScale(grey_img,1.5,5)
x,y,w,h=face[0][0],face[0][1],face[0][2],face[0][3]
frame=im[y:y+h,x:x+w]
nose=nose_classifier.detectMultiScale(frame,1.5,5)
for (xn,yn,wn,hn) in nose:
    row,col,_=frame.shape
    h,w,_=mustache.shape
    #cv2.rectangle(im,(xn,yn),(xn+wn,yn+hn),(255,255,0),2) 
    mustache1=cv2.resize(mustache.copy(),(wn,hn))
    h,w,_=mustache1.shape
    for i in range(0,h):
        for j in range(0,w):
            if xn + j >= col or yn + i >= row:
                continue
            if mustache1[i][j][3]!=0:
                frame[int(h/2)+yn+i][xn+j]=mustache1[i][j]
                

cv2.imshow('Jamie show ',im)    
cv2.waitKey(0)
cv2.destroyAllWindows()    
```
