```python
import cv2
import matplotlib.pyplot as plt
```

```python
# read an image
im=cv2.imread('dog.jpg')
newimage=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(im)
plt.show()
plt.imshow(newimage)
plt.show()
```

```python
# plotting using cv2 libraries
cv2.imshow(' DOG ',newimage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
