```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading data
x=pd.read_csv('DATASET/Train.csv').values
X_train=x[:,:5]
Y_train=x[:,-1:]
```

```python
print(X_train.shape,Y_train.shape)
```

```python
# Appending colums of one to starting of X_train
ones=np.ones((X_train.shape[0],1))
X_train=np.concatenate((ones,X_train),axis=1)
```

```python
def initialise(X):
    w=np.zeros((1,X.shape[1]))
    print(w.shape)
    return w
def hypothesis(X,theta):
    return np.dot(theta,X.T)
def error(X,Y,theta):
    m=X.shape[0]
    hx=hypothesis(X,theta)
    return np.sum((hx-Y.T)**2)/(2*m)
def gradient_calculation(X,Y,theta):
    hx=hypothesis(X,theta)
    m=X.shape[0]
    dz=(hx-Y.T)/m
    dw=np.dot(dz,X)
    return dw
def gradient_update_rule(X,Y,w,learning_rate=0.05,max_iterations=1000):
    err=[]
    for i in range(max_iterations):
        dw=gradient_calculation(X,Y,w)
        er=error(X,Y,w)
        err.append(er)
        w=w-learning_rate*dw
        print(w.shape)
    return err,w

w=initialise(X_train)
err,w=gradient_update_rule(X_train,Y_train,w)
```

```python
plt.plot(err)
```
